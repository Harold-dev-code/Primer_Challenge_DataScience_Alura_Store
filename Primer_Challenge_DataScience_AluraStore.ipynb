# Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

#### 1. Verificar columnas de los DataFrames:

```python
print("Columnas de tienda1:", tienda1.columns.tolist())
print("Columnas de tienda2:", tienda2.columns.tolist())
print("Columnas de tienda3:", tienda3.columns.tolist())
print("Columnas de tienda4:", tienda4.columns.tolist())

```

```python
# Comprobación programática
columns_tienda1 = set(tienda1.columns)
columns_tienda2 = set(tienda2.columns)
columns_tienda3 = set(tienda3.columns)
columns_tienda4 = set(tienda4.columns)

if columns_tienda1 == columns_tienda2 == columns_tienda3 == columns_tienda4:
    print("\nTodos los DataFrames tienen las mismas columnas.")
else:
    print("\nLos DataFrames tienen columnas diferentes.")
```

#### 2. Añadir columna 'tienda' a cada DataFrame: Asignar un identificador

```python
tienda1['Tienda'] = 'tienda1'
tienda2['Tienda'] = 'tienda2'
tienda3['Tienda'] = 'tienda3'
tienda4['Tienda'] = 'tienda4'

print("Column 'Tienda' añadida a todos los DataFrames.")
```

#### 3. Concatenar todos los DataFrames:

```python
df_tiendas_completo = pd.concat([tienda1, tienda2, tienda3, tienda4], ignore_index=True)
df_tiendas_completo.head()
```

#1. Análisis de facturación

## En este primer análisis, calculamos el ingreso total de cada tienda. Sumando los valores de la columna Precio.



```python
# agrupo por la columna 'Tienda' y para cada Tienda selecciono la columna 'precio' calculando la suma de todos sus valores con .sum()
ingreso_total_por_tienda = df_tiendas_completo.groupby('Tienda')['Precio'].sum()
# El resultado es una Serie de Pandas donde el índice son los nombres de las tiendas y los valores son el ingreso total de cada tienda.
print("="*60)
print("💲💰💲Ingreso total por tienda:")
print("="*60)
# Iterar sobre cada tienda e ingreso(precio_total) para imprimirlo formateado.
for tienda, ingreso in ingreso_total_por_tienda.items():
    print(f"{tienda}: 💲{ingreso:,.2f}")
# Sumar la columna 'Precio' del DataFrame completo para obtener la facturación total.
facturacion_total = df_tiendas_completo['Precio'].sum()
# Imprimir la facturación total formateada.
print("="*60)
print(f"\n🏦💵💰Facturación total de todas las tiendas💰: 💲{facturacion_total:,.2f}")
print("="*60)
```

# 2. Ventas por categoría

##calculamos la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

```python
# Agrupa por tienda y por categoría de producto para calcular las ventas totales de cada combinación. El resultado es una Series que muestra el ingreso total por cada combinación de tienda y categoría.
ventas_por_categoria_tienda = df_tiendas_completo.groupby(['Tienda', 'Categoría del Producto'])['Precio'].sum()
print("="*60)
print("📊 Ordenamiento Global🏆:")
print("="*60)
# Aparecen las categorías que más dinero generaron en total en orden descendente, sin importar a qué tienda pertenecen
display(ventas_por_categoria_tienda.sort_values(ascending=False))
print("="*60)
print("🚀 Ordenamiento Por Tiendas")
print("="*60)
# Logra ordenar por cada Tienda y muestre sus productos del más vendido al menos.
# La función lambda (lambda x: x.sort_values(ascending=False)) se aplica a cada grupo de tienda ('x').
# Para cada tienda, 'x' representa sus ventas por categoría, y 'x.sort_values(ascending=False)' ordena esas ventas de mayor a menor.
display(ventas_por_categoria_tienda.groupby('Tienda', group_keys= False).apply(lambda x: x.sort_values(ascending=False)))
```

# 3. Calificación promedio de la tienda


##Calculamos la calificación promedio de los productos vendidos por cada tienda.

```python
# Agrupa el DataFrame por 'Tienda' y calcula la calificación promedio para cada una con .mean(), creando una Serie con estos valores.
calificacion_promedio_tienda = df_tiendas_completo.groupby('Tienda')['Calificación'].mean()
print("\n⭐ RANKING DE SATISFACCIÓN ⭐")
print("------------------------------------------")
print("║  🏆  TOP TIENDAS POR CALIFICACIÓN PROMEDIO     ║")
print("📌 Nota: Ordenado de mayor a menor puntaje")
# Muestra la serie de calificaciones promedio por tienda, ordenada de mayor a menor.
display(calificacion_promedio_tienda.sort_values(ascending=False))
```

# 4. Productos más y menos vendidos

```python
# Identificar los productos más y menos vendidos en general.
```

```python
productos_mas_vendidos = df_tiendas_completo['Producto'].value_counts().head(5)
print("Los 5 productos más vendidos:")
display(productos_mas_vendidos)

productos_menos_vendidos = df_tiendas_completo['Producto'].value_counts().tail(5)
print("\nLos 5 productos menos vendidos:")
display(productos_menos_vendidos)
```

# 5. Envío promedio por tienda

```python
# Calcular el costo de envío promedio por tienda.
```

```python
envio_promedio_tienda = df_tiendas_completo.groupby('Tienda')['Costo de envío'].mean()
print("Costo de envío promedio por tienda:")
display(envio_promedio_tienda.sort_values(ascending=False))
```
