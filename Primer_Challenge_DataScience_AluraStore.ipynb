# Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

para esto debo realizar lo siguiente:

3. Concatenar todos los DataFrames: Unir todos los DataFrames en uno solo, incluyendo la nueva columna 'tienda'.

#### 1. Verificar columnas de los DataFrames:

```python
print("Columnas de tienda1:", tienda1.columns.tolist())
print("Columnas de tienda2:", tienda2.columns.tolist())
print("Columnas de tienda3:", tienda3.columns.tolist())
print("Columnas de tienda4:", tienda4.columns.tolist())

```

```python
# Comprobación programática
columns_tienda1 = set(tienda1.columns)
columns_tienda2 = set(tienda2.columns)
columns_tienda3 = set(tienda3.columns)
columns_tienda4 = set(tienda4.columns)

if columns_tienda1 == columns_tienda2 == columns_tienda3 == columns_tienda4:
    print("\nTodos los DataFrames tienen las mismas columnas.")
else:
    print("\nLos DataFrames tienen columnas diferentes.")
```

#### 2. Añadir columna 'tienda' a cada DataFrame: Asignar un identificador

```python
tienda1['Tienda'] = 'tienda1'
tienda2['Tienda'] = 'tienda2'
tienda3['Tienda'] = 'tienda3'
tienda4['Tienda'] = 'tienda4'

print("Column 'Tienda' añadida a todos los DataFrames.")
```

#### 3. Concatenar todos los DataFrames:

```python
df_tiendas_completo = pd.concat([tienda1, tienda2, tienda3, tienda4], ignore_index=True)
df_tiendas_completo.head()
```

#### 4. Comprobar que los dataframe se unieron correctamente

##### con .sample(n=)

```python
df_tiendas_completo.sample(n=7)
```

##### Comprobar tambien sumando las filas de todos los dataframe utilizando:   .shape[0] y comparandolo con el shape[0] del completo.

```python
total_rows = tienda1.shape[0] + tienda2.shape[0] + tienda3.shape[0] + tienda4.shape[0]
print(f"El número total de filas entre todas los df de las tiendas es: {total_rows}")
```

```python
df_tiendas_completo.shape[0]
print(f"El número total de filas del df_tiendas_completo es: {df_tiendas_completo.shape[0]}")
```
