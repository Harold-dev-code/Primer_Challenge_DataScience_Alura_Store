#5. **Análisis de facturación**

## En este primer análisis, calculamos el ingreso total de cada tienda. Sumando los valores de la columna Precio.



```python
# agrupo por la columna 'Tienda' y para cada Tienda selecciono la columna 'precio' calculando la suma de todos sus valores con .sum()
ingreso_total_por_tienda = df_tiendas_completo.groupby('Tienda')['Precio'].sum()
# El resultado es una Serie de Pandas donde el índice son los nombres de las tiendas y los valores son el ingreso total de cada tienda.
print("="*60)
print("💲💰💲Ingreso total por tienda:")
print("="*60)
# Iterar sobre cada tienda e ingreso(precio_total) para imprimirlo formateado.
for tienda, ingreso in ingreso_total_por_tienda.items():
    print(f"{tienda}: 💲{ingreso:,.2f}")
# Sumar la columna 'Precio' del DataFrame completo para obtener la facturación total.
facturacion_total = df_tiendas_completo['Precio'].sum()
# Imprimir la facturación total formateada.
print("="*60)
print(f"\n🏦💵💰Facturación total de todas las tiendas💰: 💲{facturacion_total:,.2f}")
print("="*60)
```

#6. **Ventas por categoría**

##calculamos la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

```python
# Agrupa por tienda y por categoría de producto para calcular las ventas totales de cada combinación. El resultado es una Series que muestra el ingreso total por cada combinación de tienda y categoría.
ventas_por_categoria_tienda = df_tiendas_completo.groupby(['Tienda', 'Categoría del Producto'])['Precio'].sum()
print("="*60)
print("📊 Ordenamiento Global🏆:")
print("="*60)
# Aparecen las categorías que más dinero generaron en total en orden descendente, sin importar a qué tienda pertenecen
display(ventas_por_categoria_tienda.sort_values(ascending=False).map(lambda x: f'💲{x:,.2f}'))
print("="*60)
print("🚀 Ordenamiento Por Tiendas")
print("="*60)
# Logra ordenar por cada Tienda y muestre sus productos del más vendido al menos.
# La función lambda (lambda x: x.sort_values(ascending=False)) se aplica a cada grupo de tienda ('x').
# Para cada tienda, 'x' representa sus ventas por categoría, y 'x.sort_values(ascending=False)' ordena esas ventas de mayor a menor.
display(ventas_por_categoria_tienda.groupby('Tienda', group_keys= False).apply(lambda x: x.sort_values(ascending=False)).map(lambda x: f'💲{x:,.2f}'))
```

#7. **Calificación promedio de la tienda**


##calculamos las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.

```python
# Agrupa el DataFrame por 'Tienda' y calcula la calificación promedio para cada una con .mean(), creando una Serie con estos valores.
calificacion_promedio_tienda = df_tiendas_completo.groupby('Tienda')['Calificación'].mean()
print("\n⭐ RANKING DE SATISFACCIÓN ⭐")
print("------------------------------------------")
print("║  🏆  TOP TIENDAS POR CALIFICACIÓN PROMEDIO     ║")
print("📌 Nota: Ordenado de mayor a menor puntaje")
# Muestra la serie de calificaciones promedio por tienda, ordenada de mayor a menor.
display(calificacion_promedio_tienda.sort_values(ascending=False).map(lambda x: f'{x:.2f}'))
```

#8. **Productos más y menos vendidos**

##Identificar los productos más y menos vendidos en general.


```python
print("="*60)
# Calcula y imprime el número total de productos únicos en el DataFrame 'df_tiendas_completo' utilizando:
# .nunique() metodo de pandas para contar ignorando duplicados y valores nulos .
print(f"\nNúmero total de productos diferentes: {df_tiendas_completo['Producto'].nunique()}")
print("="*60)
# Calcula la frecuencia de cada producto (cuántas veces aparece)con:_
# value_counts()nos da un desglose detallado de cuántas veces aparece cada elemento único.
productos_mas_vendidos = df_tiendas_completo['Producto'].value_counts()
print("="*60)
print("\nVision General de todos los productos\ndesde el más vendido al menos vendido(Todas Las Tiendas):")
print("="*60)
display(productos_mas_vendidos)
```

##Identificar los que 3 que generaron más Dinero y los que 3 que generaron menos Dinero

```python
# Agregamos por Producto para obtener el recaudo total y la cantidad vendida usando el metodo:
#.agg()
'''que se usa para aplicar una o más operaciones de agregación (como suma, promedio, conteo, máximo, mínimo, etc.)
a uno o más columnas de un DataFrame o de un grupo de DataFrames (cuando se usa después de groupby).
Es muy potente para resumir datos.'''

df_productos_agg = df_tiendas_completo.groupby('Producto').agg(
    total_recaudo=('Precio', 'sum'), #  se crea nueva_columna = ('columna_original', 'alias_funcion_agregacion'('sum'))
    cantidad_vendida=('Producto', 'count') #se crea nueva_columna = ('columna_original', 'alias_funcion_agregacion'('count'))
)
#se crea dataframe 'productos_con_mas_recaudo' con Los 3 productos con más recaudo tomando el DataFrame
#df_productos_agg (que ya tiene el recaudo total por producto) y con sort_values(by='total_recaudo', ascending=False)
#se ordenan los productos de mayor a menor según su total_recaudo.
productos_con_mas_recaudo = df_productos_agg.sort_values(by='total_recaudo', ascending=False).head(3) #.head(3): Selecciona los 3 primeros

print("="*60)
print("Los 3 productos con más recaudo (Todas Las Tiendas):")
print("="*60)
#El bucle for imprime estos 3 productos junto con su cantidad vendida y el recaudo total, de forma legible.
for producto, row in productos_con_mas_recaudo.iterrows():
    print(f"{producto}: Se vendieron {int(row['cantidad_vendida'])} Unidades y se Recaudo 💲{row['total_recaudo']:,.2f} ")
    #int(...): Fuerza a que ese valor sea tratado como un número entero.
'''iterrows() devuelve en cada iteración una tupla del DF productos_con_mas_recaudo, compuesta por
   (indice de la fila, la fila como una Serie). y con for producto, row in ...,
   estás "desempaquetando" esa tupla para guardar el índice en producto y la Serie de la fila en row.'''

# Los 3 productos con menos recaudo (ordenados del menos al más rentable de los 3)
# se usa la misma logica pero se cambia ascending=True para ordenar de menor a mayor recaudo.
productos_con_menos_recaudo = df_productos_agg.sort_values(by='total_recaudo', ascending=True).head(3)
print("="*60)
print("\nLos 3 productos con menos recaudo (Todas Las Tiendas):")
print("="*60)
for producto, row in productos_con_menos_recaudo.iterrows():
    print(f"{producto}: Se vendieron {int(row['cantidad_vendida'])} Unidades y se Recaudo 💲{row['total_recaudo']:,.2f}")
print("="*60)
```

##mostrar por tienda los productos que más recaudaron dinero y los que menos

```python

print("="*60)
print("📊 Top 3 Productos con más recaudo por Tienda 🏆")
print("="*60)

# Agrupamos el DataFrame completo por 'Tienda' y 'Producto' y suma el 'Precio'.
# Esto calcula el recaudo total para cada producto dentro de cada tienda.
recaudo_por_producto_por_tienda = df_tiendas_completo.groupby(['Tienda', 'Producto'])['Precio'].sum()
# Para cada tienda, identificamos los 3 productos que generaron más recaudo.
# 'groupby(level=0)' agrupa por el primer nivel del índice (que es 'Tienda').
# '.apply(lambda x: x.nlargest(3))' selecciona los 3 valores más grandes (recaudo) dentro de cada grupo.
top_3_productos_por_tienda = recaudo_por_producto_por_tienda.groupby(level=0).apply(lambda x: x.nlargest(3))
# Inicializa una variable para rastrear la tienda actual y evitar imprimir el encabezado repetidamente.
current_tienda = None
# Itera sobre los resultados de los 3 productos con más recaudo por tienda.
# 'items()' devuelve pares de (índice, valor), donde el índice es una tupla (Tienda, Producto).
for index_tuple, precio in top_3_productos_por_tienda.items():
    # Extrae el nombre de la tienda del índice de la tupla.
    tienda = index_tuple[0]
    # Extrae el nombre del producto del índice de la tupla.
    producto = index_tuple[2]
    # Comprueba si la tienda actual es diferente de la tienda procesada anteriormente.
    if tienda != current_tienda:
        # Si es una nueva tienda, imprime su nombre y una línea de separación.
        print(f"\nTienda: {tienda}")
        print("-------------------------")
        # Actualiza la tienda actual.
        current_tienda = tienda
    # Imprime el producto y su recaudo formateado.
    # '{producto:<25}' formatea el nombre del producto para ocupar 25 caracteres, alineado a la izquierda.
    print(f"  Producto: {producto:<25} Recaudo: 💲{precio:,.2f}")
# Imprime una línea de separación larga al final de esta sección.
print("="*80)

# Imprime una línea de separación y el título para los productos con menos recaudo.
print("="*60)
print("📊 Top 3 Productos con menos recaudo por Tienda 🏆")
print("="*60)

# Para cada tienda, identifica los 3 productos que generaron menos recaudo.
# 'nsmallest(3)' selecciona los 3 valores más pequeños (recaudo) dentro de cada grupo.
inf_3_productos_por_tienda= recaudo_por_producto_por_tienda.groupby(level=0).apply(lambda x: x.nsmallest(3))

# Reinicia la variable de la tienda actual para la nueva sección.
current_tienda = None
# Itera sobre los resultados de los 3 productos con menos recaudo por tienda.
for index_tuple, precio in inf_3_productos_por_tienda.items():
  # Extrae el nombre de la tienda del índice de la tupla.
  tienda = index_tuple[0]
  # Extrae el nombre del producto del índice de la tupla.
  producto = index_tuple[2]
  # Comprueba si la tienda actual es diferente de la tienda procesada anteriormente.
  if tienda != current_tienda:
    # Si es una nueva tienda, imprime su nombre y una línea de separación.
    print(f"\nTienda: {tienda}")
    print("-------------------------")
    # Actualiza la tienda actual.
    current_tienda = tienda
  # Imprime el producto y su recaudo formateado.
  print(f"  Producto: {producto:<25} Recaudo: 💲{precio:,.2f}")
# Imprime una línea de separación larga al final de esta sección.
print("="*80)
```

## Ventas totales por tienda

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import FuncFormatter #herramienta específica que te permite crear una función personalizada
#para decidir cómo se escribe el texto en los ejes (por ejemplo, ponerle el signo de pesos, comas de miles, etc.).

# Configura el estilo visual de las gráficas (fondo blanco con cuadrícula)
sns.set_style("whitegrid")

# Define el tamaño de la ventana de la gráfica (ancho: 8 pulgadas, alto: 5 pulgadas)
plt.figure(figsize=(8, 5))

# Determina cuántas tiendas hay en el conjunto de datos
num_tiendas = len(ingreso_total_por_tienda)

# Crea una lista de colores inicializada con azul claro para todas las barras
colores = ['skyblue'] * num_tiendas

# Localiza el índice del valor más alto y lo pinta de verde
colores[ingreso_total_por_tienda.values.argmax()] = 'green'

# Localiza el índice del valor más bajo y lo pinta de rojo
colores[ingreso_total_por_tienda.values.argmin()] = 'red'

# Genera el gráfico de barras asignando datos, colores y eliminando la leyenda redundante
ax = sns.barplot(
    x=ingreso_total_por_tienda.index, # Asigna los nombres de las tiendas al eje X.
    y=ingreso_total_por_tienda.values, # Asigna los montos de ingreso total al eje Y.
    palette=colores, # Aplica la lista de colores personalizada a las barras.
    hue=ingreso_total_por_tienda.index, # Se usa para que Seaborn asigne colores basándose en el índice (nombres de tienda).
    legend=False # Deshabilita la leyenda, ya que los colores tienen un significado directo en el gráfico.
)

# Establece el título principal, aumentando el tamaño y resaltando en negrita
plt.title('Ventas Totales por Tienda', fontsize=16, fontweight='bold')

# Etiqueta el eje horizontal (nombres de las tiendas)
plt.xlabel('Tienda', fontsize=12,fontweight='bold')

# Etiqueta el eje vertical (monto de ingresos)
plt.ylabel('Ingreso Total', fontsize=12,fontweight='bold')

# Aplica un formato de moneda al eje Y (agrega comas como separadores de miles y 0 decimales)
ax.yaxis.set_major_formatter(FuncFormatter(lambda x, p: f'{x:,.0f}'))

# Rota los nombres de las tiendas 45 grados para que no se amontonen y los alinea a la derecha
plt.xticks(rotation=45, ha='right' )

# Ajusta los márgenes automáticamente para que no se corte ninguna etiqueta
plt.tight_layout()

# Renderiza y muestra el gráfico final en pantalla
plt.show()
```

## Ventas_Anuales_y_Clasificación_por_Tienda

Calcular las ventas totales de cada tienda por año (comenzando desde el primer año) y determinar su clasificación para cada año respectivo para mostrar su posición a lo largo del tiempo.


#### modificamos la columna 'fecha de compra' para cambiar el tipo dato ya que estaba como object, entonces lo pasamos a datetime con la función pd.to_datatime.


```python
# Convierte la columna 'Fecha de Compra' a formato de fecha y hora (datetime)
df_tiendas_completo['Fecha de Compra'] = pd.to_datetime(df_tiendas_completo['Fecha de Compra'], format='%d/%m/%Y')
df_tiendas_completo['Año'] = df_tiendas_completo['Fecha de Compra'].dt.year # Extrae el año de la columna 'Fecha de Compra' y lo asigna a una nueva columna 'Año'.
ventas_anuales_por_tienda = df_tiendas_completo.groupby(['Año', 'Tienda'])['Precio'].sum() # Agrupa los datos por 'Año' y 'Tienda', y luego suma los 'Precio' para obtener las ventas anuales por tienda.

print("Ventas anuales por tienda:\n") # Imprime un título para la sección de ventas anuales.
display(ventas_anuales_por_tienda.map(lambda x: f'💲{x:,.2f}')) # Muestra las ventas anuales por tienda, formateando los valores como moneda.
```

## Graficamos




```python
import matplotlib.pyplot as plt # Importa la librería Matplotlib para la creación de gráficos.
import seaborn as sns # Importa la librería Seaborn para mejorar el estilo de los gráficos.
from matplotlib.ticker import FuncFormatter # Importa FuncFormatter para crear un formateador personalizado para los ejes.

sns.set_style("whitegrid") # Configura el estilo visual de los gráficos a un fondo blanco con cuadrícula.

plt.figure(figsize=(10, 6)) # Crea una nueva figura para el gráfico con un tamaño específico (ancho x alto).

# Asegurarse de que 'ventas_anuales_df' esté definido y contenga los rankings para el gráfico.
ventas_anuales_df = ventas_anuales_por_tienda.reset_index() # Convierte la Serie 'ventas_anuales_por_tienda' en un DataFrame y reinicia su índice.
ventas_anuales_df['Ranking'] = ventas_anuales_df.groupby('Año')['Precio'].rank(ascending=False, method='min') # Calcula el ranking de ventas ('Precio') por 'Año' y lo añade como una nueva columna 'Ranking'.

# Obtener tiendas únicas para iterar y trazar una línea por cada una.
unidades_tienda = ventas_anuales_df['Tienda'].unique() # Obtiene una lista de los nombres de tiendas únicas.

# Graficar una línea para cada tienda, mostrando su evolución en el ranking.
for tienda in unidades_tienda: # Itera sobre cada nombre de tienda.
    tienda_data = ventas_anuales_df[ventas_anuales_df['Tienda'] == tienda] # Filtra el DataFrame para obtener los datos de la tienda actual.
    plt.plot(tienda_data['Año'], tienda_data['Ranking'], marker='o', label=tienda) # Traza una línea con marcadores para el ranking de la tienda a lo largo de los años.

# Establecer título y etiquetas de los ejes del gráfico.
plt.title('Evolución del Ranking de Ventas Anuales por Tienda (Top)', fontsize=16, fontweight='bold') # Establece el título principal del gráfico.
plt.xlabel('Año', fontsize=12, fontweight='bold') # Etiqueta el eje X como 'Año'.
plt.ylabel('Ranking de Ventas (1° es mejor)', fontsize=12, fontweight='bold') # Etiqueta el eje Y como 'Ranking de Ventas'.

# Invertir el eje Y para que el ranking #1 (el mejor) aparezca en la parte superior del gráfico.
plt.gca().invert_yaxis() # Invierte el eje Y para que los valores más bajos (mejor ranking) estén arriba.

# Establecer los 'ticks' del eje Y para mostrar solo números enteros desde 1 hasta el ranking máximo.
max_rank = ventas_anuales_df['Ranking'].max() # Obtiene el valor máximo del ranking.
plt.yticks(range(1, int(max_rank) + 1)) # Configura los 'ticks' del eje Y para mostrar enteros.

# --- Formato personalizado para las etiquetas del eje Y (ordinales) ---
def ordinal_formatter(x, p): # Define una función para formatear los 'ticks' del eje Y como números ordinales.
    # 'x' es el valor numérico del 'tick' (por ejemplo, 1, 2, 3).
    # 'p' es la posición del 'tick' en el eje (no se usa en este caso).
    return f'{int(x)}°'    # Convierte el valor 'x' a un entero y le añade el símbolo de grado '°' para formar el ordinal (ej. '1°').

plt.gca().yaxis.set_major_formatter(FuncFormatter(ordinal_formatter)) # Aplica el formateador ordinal al eje Y. .set_major_formatter() método que nos permite definir cómo se van a mostrar las etiquetas principales en el eje.
# ------------------------------------------------------------------

# Añadir una leyenda para identificar cada línea de tienda en el gráfico.
plt.legend(title='Tienda') # Muestra la leyenda, usando los nombres de las tiendas.

# Ajustar el diseño del gráfico para evitar que las etiquetas se superpongan.
plt.tight_layout() # Ajusta automáticamente los parámetros de la subtrama para un diseño ajustado.

# Mostrar el gráfico final en pantalla.
plt.show() # Muestra el gráfico.
```

##Buscar si el problema de las ventas de la tienda 4 deriva de la mala atención por parte de los vendedores.

```python
# Agrupa el DataFrame completo por 'Vendedor' y 'Tienda', calcula la calificación promedio y reinicia el índice.
#calcula la calificación promedio que cada vendedor obtuvo en cada tienda específica donde realizó ventas.
df_vendedor_calificacion = df_tiendas_completo.groupby(['Vendedor', 'Tienda'])['Calificación'].mean().reset_index()
#print(df_vendedor_calificacion)
# Renombra la columna 'Calificación' a 'Calificacion_Promedio_Ventas' para mayor claridad.
df_vendedor_calificacion = df_vendedor_calificacion.rename(columns={'Calificación': 'Calificacion_Promedio_Ventas'})

# Agrupa el DataFrame 'df_vendedor_calificacion' por 'Vendedor', calcula el promedio de sus 'Calificacion_Promedio_Ventas' y reinicia el índice.
#toma los resultados de la primera agrupación (es decir, las calificaciones promedio que un vendedor obtuvo en cada tienda) y luego calcula un promedio de esos promedios.
calificacion_promedio_general_vendedor = df_vendedor_calificacion.groupby('Vendedor')['Calificacion_Promedio_Ventas'].mean().reset_index()
# Renombra la columna 'Calificacion_Promedio_Ventas' a 'Calificacion_Promedio_General'.
calificacion_promedio_general_vendedor = calificacion_promedio_general_vendedor.rename(columns={'Calificacion_Promedio_Ventas': 'Calificacion_Promedio_General'})
#print(calificacion_promedio_general_vendedor)
print("Calificación promedio general de los vendedores (en todas las tiendas):")
# Muestra el DataFrame 'calificacion_promedio_general_vendedor', ordenado por la calificación general de forma descendente.
# Además, utiliza '.style.format()' para formatear la columna 'Calificacion_Promedio_General' con dos decimales.
display(calificacion_promedio_general_vendedor.sort_values(by='Calificacion_Promedio_General', ascending=False).set_index('Vendedor').style.format({'Calificacion_Promedio_General': '{:.2f}'}))
```

```python
from matplotlib.typing import LineStyleType
import matplotlib.pyplot as plt # Importa la librería Matplotlib para la creación de gráficos.
import seaborn as sns # Importa Seaborn para mejorar el estilo de los gráficos y facilitar la visualización.
import pandas as pd # Importa Pandas para manipulación y análisis de datos en DataFrames.

sns.set_style("whitegrid") # Configura el estilo visual de los gráficos a un fondo blanco con cuadrícula.

# Define una función para categorizar las calificaciones promedio en rangos de color específicos.
def get_rating_color_category(rating):
    if rating > 4: # Si la calificación es mayor a 4
        return 'Excelente' # Categoriza como 'Excelente'
    elif 3 <= rating <= 4: # Si la calificación está entre 3 y 4 (inclusive)
        return 'Bueno' # Categoriza como 'Bueno'
    elif 2 <= rating < 3: # Si la calificación está entre 2 y menos de 3
        return 'Regular' # Categoriza como 'Regular'
    elif 1 <= rating < 2: # Si la calificación está entre 1 y menos de 2
        return 'Malo' # Categoriza como 'Malo'
    else: # Para cualquier otra calificación (menos de 1)
        return 'Muy Malo' # Categoriza como 'Muy Malo'

# Aplica la función 'get_rating_color_category' a la columna de calificación promedio
# y guarda el resultado en una nueva columna 'Rating_Color_Category'.
calificacion_promedio_general_vendedor['Rating_Color_Category'] = calificacion_promedio_general_vendedor['Calificacion_Promedio_General'].apply(get_rating_color_category)

# Define una paleta de colores personalizada para cada categoría de calificación.
custom_palette = {
    'Excelente': 'green', # Verde para calificaciones excelentes
    'Bueno': 'lightgreen', # Verde claro para calificaciones buenas
    'Regular': 'orange', # Naranja para calificaciones regulares
    'Malo': 'red', # Rojo para calificaciones malas
    'Muy Malo': 'darkred' # Rojo oscuro para calificaciones muy malas
}

# Define el orden de las categorías para asegurar una leyenda consistente.
category_order = ['Excelente', 'Bueno', 'Regular', 'Malo', 'Muy Malo']
# Convierte la nueva columna de categoría de color a un tipo categórico con el orden definido.
calificacion_promedio_general_vendedor['Rating_Color_Category'] = pd.Categorical(calificacion_promedio_general_vendedor['Rating_Color_Category'], categories=category_order, ordered=True)

plt.figure(figsize=(14, 7)) # Crea una nueva figura para el gráfico con un tamaño específico (ancho: 14, alto: 7 pulgadas).

sns.scatterplot(
    x='Vendedor', # Asigna los nombres de los vendedores al eje X.
    y='Calificacion_Promedio_General', # Asigna la calificación promedio general al eje Y.
    data=calificacion_promedio_general_vendedor, # Especifica el DataFrame a usar.
    hue='Rating_Color_Category', # Usa la columna de categorías de color para asignar el color de cada punto.
    s=200, # Establece el tamaño de los puntos en el gráfico.
    palette=custom_palette, # Aplica la paleta de colores personalizada.
    legend='full' # Muestra la leyenda completa.
)

plt.title('Calificación Promedio General de Vendedores por Rango', fontsize=18, fontweight='bold') # Establece el título principal del gráfico.
plt.xlabel('Vendedor', fontsize=14, fontweight='bold') # Etiqueta el eje X.
plt.ylabel('Calificación Promedio', fontsize=16, fontweight='bold') # Etiqueta el eje Y.
plt.xticks(rotation=60, ha='right', fontsize=10) # Rota las etiquetas del eje X para mejorar la legibilidad y ajusta su tamaño.
plt.yticks(fontsize=14) # Ajusta el tamaño de fuente de las etiquetas del eje Y.
plt.ylim(0, 5) # Establece los límites del eje Y de 0 a 5, correspondiente a la escala de calificaciones.
plt.grid(True, linestyle='--', alpha=0.9 ) # Añade una cuadrícula al gráfico con estilo de línea discontinua y transparencia.

# Ajusta la posición de la leyenda a la derecha del gráfico y le añade un marco.
plt.legend(title='Rango de Calificación', bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0., frameon=True)

# Ajusta el diseño del gráfico para evitar que las etiquetas o la leyenda se superpongan, reservando espacio para la leyenda.
plt.tight_layout(rect=[0, 0, 0.85, 1])
plt.show() # Muestra el gráfico final.
```

## Generamos gráficos de torta para visualizar la distribución de ventas de las cuatro tiendas en cada ciudad.

```python
import matplotlib.pyplot as plt
import seaborn as sns # Importar seaborn para la paleta de colores
from matplotlib.ticker import FuncFormatter

# Función para generar gráficos de torta por ciudad
def plot_sales_pie_chart(city_name, df, title_prefix):
    # Filtrar el DataFrame para la ciudad específica
    df_city = df[df['Lugar de Compra'] == city_name]

    # Agrupar por tienda y sumar los precios para obtener las ventas por tienda en esa ciudad
    sales_by_store = df_city.groupby('Tienda')['Precio'].sum()

    # Calcular el porcentaje de cada tienda
    total_sales_city = sales_by_store.sum()

    # Asegurarse de que haya datos para graficar
    if total_sales_city == 0:
        print(f"No hay datos de ventas para {city_name}.")
        return

    percentages = (sales_by_store / total_sales_city) * 100

    # Preparar datos para el gráfico de torta
    labels = percentages.index.tolist()
    sizes = percentages.values.tolist()

    # Crear el gráfico de torta
    plt.figure(figsize=(8, 8))
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
    plt.title(f'{title_prefix} {city_name} (Total: $ {total_sales_city:,.2f})', fontsize=16, fontweight='bold')
    plt.axis('equal') # Asegura que el gráfico de torta sea circular.
    plt.show()

# Obtener la lista de todas las ciudades únicas
todas_las_ciudades = df_tiendas_completo['Lugar de Compra'].unique().tolist()

# Generar gráficos de torta para cada ciudad
for ciudad in todas_las_ciudades:
    plot_sales_pie_chart(ciudad, df_tiendas_completo, 'Distribución de Ventas en')
```
