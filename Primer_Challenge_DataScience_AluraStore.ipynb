# Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

#### 1. Verificar columnas de los DataFrames:

```python
print("Columnas de tienda1:", tienda1.columns.tolist())
print("Columnas de tienda2:", tienda2.columns.tolist())
print("Columnas de tienda3:", tienda3.columns.tolist())
print("Columnas de tienda4:", tienda4.columns.tolist())

```

```python
# Comprobación programática
columns_tienda1 = set(tienda1.columns)
columns_tienda2 = set(tienda2.columns)
columns_tienda3 = set(tienda3.columns)
columns_tienda4 = set(tienda4.columns)

if columns_tienda1 == columns_tienda2 == columns_tienda3 == columns_tienda4:
    print("\nTodos los DataFrames tienen las mismas columnas.")
else:
    print("\nLos DataFrames tienen columnas diferentes.")
```

#### 2. Añadir columna 'tienda' a cada DataFrame: Asignar un identificador

```python
tienda1['Tienda'] = 'tienda1'
tienda2['Tienda'] = 'tienda2'
tienda3['Tienda'] = 'tienda3'
tienda4['Tienda'] = 'tienda4'

print("Column 'Tienda' añadida a todos los DataFrames.")
```

#### 3. Concatenar todos los DataFrames:

```python
df_tiendas_completo = pd.concat([tienda1, tienda2, tienda3, tienda4], ignore_index=True)
df_tiendas_completo.head()
```

#### 4. Comprobar que los dataframe se unieron correctamente

##### con .sample(n=)

```python
df_tiendas_completo.sample(n=7)
```

##### Comprobar tambien sumando las filas de todos los dataframe utilizando:   .shape[0] y comparandolo con el shape[0] del completo.

```python
total_rows = tienda1.shape[0] + tienda2.shape[0] + tienda3.shape[0] + tienda4.shape[0]
print(f"El número total de filas entre todas los df de las tiendas es: {total_rows}")
```

```python
df_tiendas_completo.shape[0]
print(f"El número total de filas del df_tiendas_completo es: {df_tiendas_completo.shape[0]}")
```

#1. Análisis de facturación

En este primer análisis, calculamos el ingreso total de cada tienda. Sumando los valores de la columna Precio.



```python
# agrupo por la columna 'Tienda' y para cada Tienda selecciono la columna 'precio' calculando la suma de todos sus valores con .sum()
ingreso_total_por_tienda = df_tiendas_completo.groupby('Tienda')['Precio'].sum()
# El resultado es una Serie de Pandas donde el índice son los nombres de las tiendas y los valores son el ingreso total de cada tienda.
print("Ingreso total por tienda:")

# Iterar sobre cada tienda e ingreso para imprimirlo formateado.
for tienda, ingreso in ingreso_total_por_tienda.items():
    print(f"{tienda}: $ {ingreso:,.2f}")
# Sumar la columna 'Precio' del DataFrame completo para obtener la facturación total.
facturacion_total = df_tiendas_completo['Precio'].sum()
# Imprimir la facturación total formateada.
print(f"\nFacturación total de todas las tiendas: ${facturacion_total:,.2f}")
```

# 2. Ventas por categoría

```python
# Calcular las ventas totales por categoría de producto.
```

```python
ventas_por_categoria = df_tiendas_completo.groupby('Categoría del Producto')['Precio'].sum()
print("Ventas totales por categoría de producto:")
display(ventas_por_categoria.sort_values(ascending=False))
```

# 3. Calificación promedio de la tienda


```python
# Calcular la calificación promedio de los productos vendidos por cada tienda.
```

```python
calificacion_promedio_tienda = df_tiendas_completo.groupby('Tienda')['Calificación'].mean()
print("Calificación promedio de productos por tienda:")
display(calificacion_promedio_tienda.sort_values(ascending=False))
```

# 4. Productos más y menos vendidos

```python
# Identificar los productos más y menos vendidos en general.
```

```python
productos_mas_vendidos = df_tiendas_completo['Producto'].value_counts().head(5)
print("Los 5 productos más vendidos:")
display(productos_mas_vendidos)

productos_menos_vendidos = df_tiendas_completo['Producto'].value_counts().tail(5)
print("\nLos 5 productos menos vendidos:")
display(productos_menos_vendidos)
```

# 5. Envío promedio por tienda

```python
# Calcular el costo de envío promedio por tienda.
```

```python
envio_promedio_tienda = df_tiendas_completo.groupby('Tienda')['Costo de envío'].mean()
print("Costo de envío promedio por tienda:")
display(envio_promedio_tienda.sort_values(ascending=False))
```
